```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys

options = webdriver.ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options = webdriver.ChromeOptions()

global title
title=[]
global reviews
reviews=[]
global date
date=[]
global rating
rating=[]
global names
names=[]
def getmain(url,name):
    options = webdriver.ChromeOptions()
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(options=options)

    driver.get(url)
    time.sleep(2)
    def datacall():
        scroll_count = 3 
        for _ in range(scroll_count):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
            time.sleep(2)  # Adjust the sleep time as needed
        time.sleep(1)
        element = driver.find_element(By.XPATH, '//*[contains(@class, "Ignyf _S Z")]')
        driver.execute_script("arguments[0].scrollIntoView();", element)
        driver.execute_script("window.scrollBy(0, -500);")
        data = driver.find_elements(By.XPATH, '//div[@data-reviewid]')
        element = driver.find_element(By.XPATH, '//*[contains(@class, "lszDU")]')
        review_id = data[0].get_attribute('data-reviewid')
   
        readmore = f'//div[@data-reviewid="{review_id}"]//*[contains(@class,"lszDU")]'
        
        readmoreclick = driver.find_element(By.XPATH, readmore)
        readmoreclick.click()

        data = driver.find_elements(By.XPATH, '//div[@data-reviewid]')
        return data

    
    


    
    def getinfo(data,name):
        for i in data:
            title.append(i.text.split('\n')[0])
    
        temp=[]
        for i in data:
            temp.append(i.text.split('Read less')[0].split('\n')[1:])
    
        for i in temp:
            t=''
            for j in i:
                t+=j
            reviews.append(t)
   
        for i in data:
            date.append(i.text.split("Date of stay:")[1].split("Trip type")[0])
            names.append(name)
        
        
    def getrating(data):
        temp=[]
        for review in data:
            review_id = review.get_attribute('data-reviewid')
   
            rating_xpath = f'//div[@data-reviewid="{review_id}"]//*[contains(@class,"ui_bubble_rating")]'
        
            rating_element = driver.find_element(By.XPATH, rating_xpath)
   
            temp.append(driver.execute_script('return window.getComputedStyle(arguments[0], "::after").getPropertyValue("content");', rating_element))
        for i in temp:
            rating.append(2*i.count('\ue00b'))

    data=datacall()
    getinfo(data,name)
    getrating(data)
    time.sleep(2)
    
    ele=49
    while ele:
        try:
            element = driver.find_element(By.XPATH, '//*[contains(@class, "ui_button nav next primary")]')
            element.click()
            
            data=datacall()
            
            getinfo(data,name)
            getrating(data)
            
            time.sleep(2)
            ele-=1
        except:
            print("didnt find")
            break
    
    
    
    
    
    
```

```python
import csv
error=[]
pname=[]
url = []
with open("D:/SwitchOff/data/Property_Names_Links.csv", mode="r", newline="", encoding="utf-8") as file:
    reader = csv.reader(file)
    for row in reader:
        if row:  
            url.append(row[3])
            pname.append(row[0])

pname=pname[1:]

url=url[1:]
```

```python
error=[]
cnt=1
for i in range(len(url)):
    try:
        getmain(url[i],pname[i])
        print("done "+ pname[i])
    except:
        print("error "+pname[i])
        error.append(pname[i])
    
    time.sleep(2)
    print(cnt)
    cnt+=1
```

```python
data = list(zip(date, rating, title, reviews, names))

fieldnames = ["Date", "Rating", "Title", "Reviews", "Property Name"]
csv_file='reviews_dataset.csv'
with open(csv_file, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)

    
    writer.writerow(fieldnames)

    for row in data:
        writer.writerow(row)

print(f"CSV file '{csv_file}' has been created.")
```
